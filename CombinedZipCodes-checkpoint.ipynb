```python
# Import the Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy.random as nr
import math

%matplotlib inline
```

```python
# Load the Dataset
AdvWorksCusts = pd.read_csv('CombiningAllZipCodes.csv', header=None)
```

```python
# Check Datatypes for each Column
AdvWorksCusts.dtypes
```

```python
AdvWorksCusts.head()
```

```python
# Make first row the header

new_header = AdvWorksCusts.iloc[0] #grab the first row for the header
AdvWorksCusts = AdvWorksCusts[1:] #take the data less the header row
AdvWorksCusts.columns = new_header #set the header row as the df header

AdvWorksCusts.head()
```

```python
# Check Datatypes for each Column
AdvWorksCusts.dtypes
```

```python
# Remove Duplicate Rows

print('Before Removing Duplicate Rows') 
print(AdvWorksCusts.shape)
AdvWorksCusts.drop_duplicates(subset = 'CustomerID', keep = 'first', inplace = True)
print('After Removing Duplicate Rows')
print(AdvWorksCusts.shape)
print(AdvWorksCusts.CustomerID.unique().shape)
```

```python
# Check Datatypes for each Column
AdvWorksCusts.dtypes
```

```python
# Drop Columns not needed

AdvWorksCusts = AdvWorksCusts.drop(['CustomerID','StateProvinceName','BirthDate1'], axis=1)
AdvWorksCusts.head()
```

```python
# Check Datatypes for each Column
AdvWorksCusts.dtypes
```

```python
# Check for Missing Values
AdvWorksCusts.isnull()
```

```python
# Check for Missing Values
AdvWorksCusts.isnull().sum()
```

```python
# Count Missing Values Per String Columns

for col in AdvWorksCusts.columns:
    if AdvWorksCusts[col].dtype == object:
        count = 0
        count = [count + 1 for x in AdvWorksCusts[col] if x == 'NaN']
        print(col + ' ' + str(sum(count)))
```

```python
AdvWorksCusts.head(20)
```

```python
# Extract the first 5 strings from the ZipCode
#sales_data["zip_code"] = sales_data["zip_code"].str[:5]
AdvWorksCusts["PostalCode"] = AdvWorksCusts["PostalCode"].str[:5]
```

```python
AdvWorksCusts.head()
```

```python
# Change Gender and MaritalStatus to 1 and 2 
code_list = [['Gender', 
              {'M' : '1', 
               'F' : '2'}],
            ['MaritalStatus', 
              {'M' : '1', 
               'S' : '2'}]]
             
for col_dic in code_list:
    col = col_dic[0]
    dic = col_dic[1]
    AdvWorksCusts[col] = [dic[x] for x in AdvWorksCusts[col]]
```

```python
# Convert Columns to Numeric

cols = ['PostalCode', 'BirthDateAge', 'HomeOwnerFlag', 'NumberCarsOwned', 'NumberChildrenAtHome',
          'TotalChildren', 'YearlyIncome']

for column in cols:
    AdvWorksCusts[column] = pd.to_numeric(AdvWorksCusts[column])
AdvWorksCusts[cols].dtypes
```
