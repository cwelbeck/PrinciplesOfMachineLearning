```python
# Import the Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy.random as nr
import math

%matplotlib inline
```

```python

# Load the Dataset
AdvWorksCusts = pd.read_csv('AdvWorksCusts.csv', header=None)
```

```python
# Check Datatypes for each Column
AdvWorksCusts.dtypes
```

```python
# Make first row the header

new_header = AdvWorksCusts.iloc[0] #grab the first row for the header
AdvWorksCusts = AdvWorksCusts[1:] #take the data less the header row
AdvWorksCusts.columns = new_header #set the header row as the df header
```

```python
# Check for duplicate rows

print(AdvWorksCusts.shape)
print(AdvWorksCusts.CustomerID.unique().shape)
```

```python
# Remove Duplicate Rows

AdvWorksCusts.drop_duplicates(subset = 'CustomerID', keep = 'first', inplace = True)
print(AdvWorksCusts.shape)
print(AdvWorksCusts.CustomerID.unique().shape)
```

```python
# The first column is customer_id, which is an identifier. We will drop this since this is not a feature.

AdvWorksCusts.drop(['CustomerID'], axis=1, inplace=True)
print(AdvWorksCusts.shape)
AdvWorksCusts.head()
```

```python
# Special Characters (Replace Special Characters)
AdvWorksCusts.columns = [str.replace('-', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace(',', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace('*', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace('/', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace('|', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace('@', '_') for str in AdvWorksCusts.columns]
AdvWorksCusts.columns = [str.replace('!', '') for str in AdvWorksCusts.columns]
```

```python
# The first column is customer_id, which is an identifier. We will drop this since this is not a feature.

AdvWorksCusts.drop(['CustomerID'], axis=1, inplace=True)
print(AdvWorksCusts.shape)
AdvWorksCusts.head()
```

```python
# Count Missing Values Per Column

for col in AdvWorksCusts.columns:
    if AdvWorksCusts[col].dtype == object:
        count = 0
        count = [count + 1 for x in AdvWorksCusts[col] if x == '?']
        print(col + ' ' + str(sum(count)))
```
